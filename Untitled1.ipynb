```python
def timing(f):
  def wrap(*args, **kwargs):
    time1=time.time()
    ret= f(*args, **kwargs)
    time2=time.time()
    print('{:s} function took {:.3f} ms'.format(f.__name__, (time2-time1)*1000.0))

    return ret
  return wrap
```

```python
@timing
def get_faces_locations(image, cnn=None):
  if cnn:
    model = "cnn"
  face_locations = face_recognition.face_locations(image, model=model)
  return face_locations
```

```python
def draw_box_faces(image, face_locations, display_image=True):
  pil_image= Image.fromarray(image)
  draw=ImageDraw.Draw(pil_image)


  for(top,right, bottomn, left) in face_locations:
    draw.rectangle(((left, top), (right, bottom)), outline=(0,0,255))
    if display_image:
      display(pil_image)
  return pil_image
```

```python
image_G7= face_recognition.load_image_file("table_ronde.jpg")

face_locations = get_faces_locations(image_G7, True)
display(draw_box_faces(image_G7, face_locations, False))
```

```python
face_encodings=face_recognition.face_encodings(image_G7, face_locations)
```

```python
known_face_names= ['Barack','françois']
known_face_encodings= [trump_encoding, trump_encoding]

people_in_new_picture =[]
for face_encoding in face_encodings:
  matches= face_recognition.compare_faces(known_face_encodings, face_encoding)
  name="Unknown"

  face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
  best_match_index= np.argmin(face_distances)
  if matches[best_match_index]:
    name= known_face_names[best_match_index]
  print(f"Found {name}")
```
